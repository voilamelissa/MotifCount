```python
import numpy as np

# A simple DNA matrix of +1/-1 match/mismatch scores
DNASIMPLE = {
    'A': {'A':1, 'C':-1, 'T':-1, 'G':-1},
    'C': {'A':-1, 'C':1, 'T':-1, 'G':-1},
    'T': {'A':-1, 'C':-1, 'T':1, 'G':-1},
    'G': {'A':-1, 'C':-1, 'T':-1, 'G':1},
}

seq1 = 'ATCG'
seq2 = 'ATTCG'

# Define our local alignment function
def localalignment(seq1,seq2):
    dim1 = len(seq1) + 1
    dim2 = len(seq2) + 1
    dp = np.zeros((dim1,dim2), dtype=int)
                
    align1 = ''
    align2 = ''
    maxscore = 0

    max_i = 0
    max_j = 0

    for i in range(1,dim1):
        for j in range(1,dim2):
            diagonal = dp[i-1,j-1] + DNASIMPLE[seq1[i-1]][seq2[j-1]]
            vertical = gap + dp[i-1,j]
            horizontal = gap + dp[i,j-1]
            dp[i,j] = max(diagonal, vertical, horizontal, 0)
    
    if dp[i,j] > maxscore:
        maxscore = dp[i,j]
        max_i = i
        max_j = j

    if dp[i,j] == 0:
        # Backtrack matrix stop
        bm[i,j] = 0
    elif dp[i,j] == diagonal:
        bm[i,j] = 1 
    elif dp[i,j] == vertical:
        bm[i,j] = 2
    elif dp[i,j] == horizontal:
        bm [i,j] = 3
    else:

i = max_i
j = max_j

while bm[i,j] != 0:

    print(dp)
    print(bm)
    print(maxscore, max_i, max_j)
    return align1, align2, maxscore

# Create a Backtracking Matrix

```
